```python
import pandas as pd

f = open("/Users/nicholasnuzhny/Downloads/new_tricks/registrations.txt", "r")

cleaned = []
for i in f:
    i = i.replace("\n", "")
    cleaned.append(i.split("<^>"))

df = pd.DataFrame(cleaned)

new_header = df.iloc[0] 
df = df[1:]
df.columns = new_header

df['registration_date'] = df['registration_date'].str[-4:]
result = df[df['package'].str.contains("Gold")]

result['registration_date'] = pd.to_numeric(result['registration_date'])

f.close()
```
